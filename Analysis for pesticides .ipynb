# Overall Pesticides Analysis 

```python
#import packages
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_columns',None)
from matplotlib.backends.backend_pdf import PdfPages
```

```python
#import data
data=pd.read_excel('capstone Project.xlsx',header=0)
data.head()
```

```python
data.shape
```

```python
# copy original data for data cleaning 
FG=data.copy()
```

# Analysis for Insecticides

```python
# sorting the data on the basis of FG-insecticides
insect= FG.loc[FG['FG'] == 'Insecticides']
insect.head()
```

```python
insect.shape
```

```python
#Sorting value on the basis of month_year and value
insect=insect[['Month_year','VALUE']]
insect.head()
```

```python
# grouping the data on the basis of month_year and value 
insect=insect.groupby('Month_year')['VALUE'].sum().reset_index()
insect
```

```python
insect.isnull().sum()
```

```python
# Set Month_year as a Index
insect=insect.set_index("Month_year")
insect
```

```python
# Check shape of the data
insect.shape
```

```python
#Resample the data on Monthly basis
y_insect=insect['VALUE'].resample('MS').mean()
y_insect 
```

```python
# check shape of the data 
y_insect.shape
```

```python
# Check if their is any missing values present 
y_insect.isnull().sum()
```

```python
# fill missing values with mean values of the given data 
y_insect=y_insect.fillna(y_insect.mean())
y_insect
```

```python
#Graph the plot 
y_insect.plot(figsize=(10,6))
plt.title("Trend Analysis for insecticide")
plt.savefig("Trend Analysis for insecticide.pdf")
plt.show()
```

```python
#Import the package and build model 
import pmdarima as pm
model_insect = pm.auto_arima(y_abc_insect,start_p=0, max_p=3, d=None, max_d=2,start_q=0, max_q=3,  
                      start_P=0,max_P=3, D=None, max_D=2, start_Q=0, max_Q=3,
                      max_order=10, m=12, seasonal=True, information_criterion='aic',
                      test='adf',trace=True,random_state=10)
model_herb = pm.auto_arima(y_abc_herb,start_p=0, max_p=3, d=None, max_d=2,start_q=0, max_q=3,  
                      start_P=0,max_P=3, D=None, max_D=2, start_Q=0, max_Q=3,
                      max_order=10, m=12, seasonal=True, information_criterion='aic',
                      test='adf',trace=True,random_state=10)
model_rodent= pm.auto_arima(y_abc_rodent,start_p=0, max_p=3, d=None, max_d=2,start_q=0, max_q=3,  
                      start_P=0,max_P=3, D=None, max_D=2, start_Q=0, max_Q=3,
                      max_order=10, m=12, seasonal=True, information_criterion='aic',
                      test='adf',trace=True,random_state=10)

model_bacter= pm.auto_arima(y_abc_bacter,start_p=0, max_p=3, d=None, max_d=2,start_q=0, max_q=3,  
                      start_P=0,max_P=3, D=None, max_D=2, start_Q=0, max_Q=3,
                      max_order=10, m=12, seasonal=True, information_criterion='aic',
                      test='adf',trace=True,random_state=10)
```

```python
print(model_insect) 
print(model_herb)
print(model_rodent)
print(model_bacter)
```

```python
# Import SARIMAX from statsmodel
from statsmodels.tsa.statespace.sarimax import SARIMAX
model_insect= SARIMAX(y_abc_insect,order=(0,0,1),seasonal_order=(0,0,3,12)).fit()
model_herb=SARIMAX(y_abc_herb,order=(1,1,1),seasonal_order=(1,0,0,12)).fit()
model_rodent=SARIMAX(y_abc_rodent,order=(0,1,2),seasonal_order=(1,0,0,12)).fit()
model_bacter=SARIMAX(y_abc_bacter,order=(0,0,0),seasonal_order=(0,0,0,0)).fit()
```

# Check how perform model or validation forecast on given data

```python
#for Insecticides 
pred_insect = model_insect.get_prediction(start=pd.to_datetime('2017-01-01'))
plt.figure(figsize=(10,6))
ax = y_abc_insect['2014':].plot(label='observed')
pred_insect.predicted_mean.plot(ax=ax, label='Validation Forecast')
ax.set_xlabel('Month_Year')
ax.set_ylabel('VALUE')
plt.legend()
plt.show()
```

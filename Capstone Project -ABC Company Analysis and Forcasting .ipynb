```python
#import packages
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_columns',None)
from matplotlib.backends.backend_pdf import PdfPages
```

```python
#import data
data=pd.read_excel('capstone Project.xlsx',header=0)
data.head()
```

```python
# copy original data for data cleaning 
abc=data.copy()
```

```python
# sorting the data on the basis of ABC Manufacturing company
abc= abc.loc[abc['COMPANY'] == 'ABC Manufacturing']
abc.head()
```

```python
abc.shape
```

```python
abc=abc[['Month_year','STATE','VALUE']]
abc.head()
```

# For Haryana

```python
# sort ABC Manufacturing data on the basis of Haryana State
abc_haryana=abc.loc[abc['STATE'] == 'Haryana']
abc_haryana.head()
```

```python
# check shape of the data 
abc_haryana.shape
```

```python
# grouping the data on the basis of month_year and value 
abc_haryana=abc_haryana.groupby('Month_year')['VALUE'].sum().reset_index()
abc_haryana
```

```python
abc_haryana.isnull().sum()
```

```python
# Set Month_year as a Index
abc_haryana=abc_haryana.set_index("Month_year")
abc_haryana
```

```python
#Resample the data on Monthly basis
y=abc_haryana['VALUE'].resample('MS').mean()
y
```

```python
# check shape of the data 
y.shape
```

```python
# Check if their is any missing values present 
y.isnull().sum()
```

```python
#y=y.replace(0,y.mean())
#y
```

```python
# fill missing values with mean values of the given data 
y=y.fillna(y.mean())
y
```

```python
#Graph the plot 
y.plot(figsize=(10,6))
plt.show()
```

```python
#Split train and test data 
train=y.loc[:'2016-12-01']
#train.tail() 
```

```python
#Split test data 
test=y.loc['2017-01-01':]
#test.tail()
```

```python
#Plotting the data 
train.plot(figsize=(10,6),title='Avg sales',fontsize=14)
test.plot(figsize=(10,6),title='Avg sales',fontsize=14)
plt.show()
```

```python
 """
Box-Jenkins Method:
1.Identification- to identify the optimum values for the parameters PDQ-auto_arima()
2.Estimation-build the model using the optimum values
3.Validation-interpret how good is the model by validation testing
4.Forecasting- forecast for future time periods
"""
```

```python
#Import the package and build model 
import pmdarima as pm
model = pm.auto_arima(y,start_p=0, max_p=3, d=None, max_d=2,start_q=0, max_q=3,  
                      start_P=0,max_P=3, D=None, max_D=2, start_Q=0, max_Q=3,
                      max_order=10, m=12, seasonal=True, information_criterion='aic',
                      test='adf',trace=True,random_state=10)
```

```python
model
```

```python
# import SARIMAX from statsmodel 
from statsmodels.tsa.statespace.sarimax import SARIMAX
model = SARIMAX(y,
                order=(1,1,1),seasonal_order=(0,0,0,12)).fit()
```

```python
# check how perform model or validation forecast on given data
pred = model.get_prediction(start=pd.to_datetime('2017-01-01'))
plt.figure(figsize=(10,6))
ax = y['2014':].plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='Validation Forecast')
ax.set_xlabel('Month_Year')
ax.set_ylabel('VALUE')
plt.legend()
plt.show()
```

```python
# check validation mean value 
pred.predicted_mean
```

```python
# check rms for validation forecast model
Y_pred = pred.predicted_mean
Y_test = y['2017-01-01':]

from sklearn.metrics import mean_squared_error
from math import sqrt
rms = sqrt(mean_squared_error(Y_test,Y_pred))
print(rms)
```

```python
# based on the validation forecast model forecast for next 12 month  
pred_uc = model.get_forecast(steps=12)
plt.figure(figsize=(10,6))
ax = y['2014':].plot(label='observed')
pred_uc.predicted_mean.plot(ax=ax, label='One-step ahead Forecast')
plt.title("ABC Manufacturing Company Forecasting2018 sales for Haryana")
ax.set_xlabel('Month_Year')
ax.set_ylabel('VALUE')
plt.legend()
plt.savefig("Forecasting2018 sales for Haryana.pdf")
plt.show()
```

```python
# forecast values for 2018
Y_predictions=pred_uc.predicted_mean
Y_predictions
```

```python
# save forecast values in excel 
Y_predictions.to_excel("ABC Manufacturing Company Forecasting2018 sales for Haryana .xlsx",index=True)
```
